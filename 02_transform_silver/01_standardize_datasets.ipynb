```python
# Load bronze tables with correct schema
customers = spark.table("project_4.bronze.customers")
orders = spark.table("project_4.bronze.orders")
order_items = spark.table("project_4.bronze.orders_items")
products = spark.table("project_4.bronze.products")

# Remove duplicates and standardize column names (example)
customers_clean = customers.dropDuplicates(["customer_id"])
orders_clean = orders.dropDuplicates(["order_id"])
order_items_clean = order_items.dropDuplicates(["order_id", "order_item_id"])
products_clean = products.dropDuplicates(["product_id"])

# Write to Silver Layer
customers_clean.write.mode("overwrite").saveAsTable("project_4.silver.customers")
orders_clean.write.mode("overwrite").saveAsTable("project_4.silver.orders")
order_items_clean.write.mode("overwrite").saveAsTable("project_4.silver.order_items")
products_clean.write.mode("overwrite").saveAsTable("project_4.silver.products")

display(customers_clean)
display(orders_clean)
display(order_items_clean)
display(products_clean)
```
