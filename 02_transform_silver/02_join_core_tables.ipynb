```python
# Load Silver tables with correct schema
customers = spark.table("project_4.silver.customers")
orders = spark.table("project_4.silver.orders")
order_items = spark.table("project_4.silver.order_items")
products = spark.table("project_4.silver.products")

# Join operations (core star schema)
sales = (
    orders
    .join(order_items, "order_id", "inner")
    .join(customers, "customer_id", "inner")
    .join(products, "product_id", "left")
)

# Write main Silver fact table
sales.write.mode("overwrite").saveAsTable("project_4.silver.sales")

display(sales)
```
