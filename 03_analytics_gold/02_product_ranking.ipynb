```python
from pyspark.sql.functions import sum, count, col

sales = spark.table("project_4.silver.sales")

product_ranking = (
    sales.groupBy("product_id", "product_category_name")
         .agg(
             count("order_id").alias("total_orders"),
             sum("price").alias("total_revenue")
         )
         .orderBy(col("total_revenue").desc())
)

product_ranking.write.mode("overwrite").saveAsTable("project_4.gold.product_ranking")

print("âœ… gold.product_ranking created successfully")
```
