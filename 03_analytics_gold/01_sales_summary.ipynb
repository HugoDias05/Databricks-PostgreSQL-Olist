```python
spark.sql("CREATE SCHEMA IF NOT EXISTS project_4.gold")
```

```python
from pyspark.sql.functions import sum, count, col

sales = spark.table("project_4.silver.sales")

sales_summary = (
    sales.groupBy("order_id", "customer_unique_id")
    .agg(
        sum(col("price").cast("float")).alias("order_total_amount"),
        count("order_item_id").alias("items_count")
    )
)

sales_summary.write.mode("overwrite").saveAsTable("project_4.gold.sales_summary")

display(sales_summary)
```
