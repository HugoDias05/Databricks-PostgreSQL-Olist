```python
from pyspark.sql.functions import to_date, sum

sales = spark.table("project_4.silver.sales")

sales_time = (
    sales.withColumn("order_date", to_date("order_purchase_timestamp"))
         .groupBy("order_date")
         .agg(sum("price").alias("daily_revenue"))
         .orderBy("order_date")
)

display(sales_time)
```
